```python
import numpy as np
import pandas as pd

class ConditionUtility:
    
    def column_exists_validator(self, cond, df):
        print('I validate stuff')
        #if column doesn't exist 
        flag = True
        if cond[0]["middle"][1:-1] not in list(df.columns):
            flag = False
        try: 
            if cond[0]["left_cond"][0] == '#':
                if cond[0]["left_cond"][1:-1] not in list(df.columns):
                    flag = False
        except:
            print()
        
        try: 
            if cond[0]["right_cond"][0] == '#':
                if cond[0]["right_cond"][1:-1] not in list(df.columns):
                    flag = False
        except:
            print()
        
        if flag:
            return {"val" : flag, "details": ""}
        else:
            return {"val" : flag, "details":"column doesn't existing"}
        
    
    #Function to parse through the various conditions, and return the value that we need to add to 
    # he column, based on whichever condition it satisfies. 
    def parse_through_condition(self,cond,df):
        #Cond is a list of conditions, from which we need to pick up each condition, 
        #and evaluate the value based on that condition
        
        msg = self.column_exists_validator(cond, df)
        if not msg["val"]:
            return msg
        
        d_mid = df[cond[0]["middle"][1:-1]]
        if df[cond[0]["middle"][1:-1]].dtype == 'object':
            d_mid = d_mid.fillna(value="")
        if df[cond[0]["middle"][1:-1]].dtype == 'float':
            d_mid = d_mid.fillna(value=0.0)
        if df[cond[0]["middle"][1:-1]].dtype == 'int':
            d_mid = d_mid.fillna(value=0)
            
        middle_vals = list(d_mid)
        
        left_vals = []
        right_vals = []
        try:
            if cond[0]["left_cond"][0] == "#":
                d_left = df[cond[0]["left_cond"][1:-1]]
                if df[cond[0]["left_cond"][1:-1]].dtype == 'object':
                    d_left = d_left.fillna(value="")
                elif df[cond[0]["left_cond"][1:-1]].dtype == 'float':
                    d_left = d_left.fillna(value=0.0)
                elif df[cond[0]["left_cond"][1:-1]].dtype == 'int':
                    d_left = d_left.fillna(value=0)
                left_vals = list(d_left)
        except:
            print('')
        
        try:
            if cond[0]["right_cond"][0] == "#":
                d_right = df[cond[0]["right_cond"][1:-1]]
                if df[cond[0]["right_cond"][1:-1]].dtype == 'object':
                    d_right = d_right.fillna(value="")
                elif df[cond[0]["right_cond"][1:-1]].dtype == 'float':
                    d_right = d_right.fillna(value=0.0)
                elif df[cond[0]["right_cond"][1:-1]].dtype == 'int':
                    d_right = d_right.fillna(value=0)
                right_vals = list(d_right)
        except:
            print('')
        
            
        final = []
            
        for i in range(len(middle_vals)):
            cnt = 0
            for j in cond:
                left_condition=j["left_cond"]
                right_condition=j["right_cond"] 
                left_sign=j["left_sign"]
                right_sign=j["right_sign"]
                
                if left_vals != []:
                    left_condition = left_vals[i]
                if right_vals != []:
                    right_condition = right_vals[i]
                
                if left_condition == "NA":
                    left_condition = ""
                if right_condition == "NA":
                    right_condition = ""
                if left_sign == "NA":
                    left_sign = ""
                if right_sign == "NA":
                    right_sign = ""
                    
                middle = middle_vals[i]
                
                
                eval_str = self.get_eval_str(left_condition, left_sign, middle, right_sign, right_condition)
    
                if eval(eval_str):
                    final.append(j["val"])
                    break;
                cnt+=1
                if cnt == len(cond):
                    final.append("")
        return final
        
    #generating string for the er    
    def get_eval_str(self, lc, ls, v, rs, rc):
        return str(lc)+str(ls)+str(v)+str(rs)+str(rc)
               
        
    #adding column to dataframe
    def add_column(self, new_col_val, name, df, position):
        
        if position["deet"] == "after":
            loc = list(df.columns).index(position["col"]) + 1
            df.insert(loc, name, new_col_val)
        elif position["deet"] == "before":
            loc = list(df.columns).index(position["col"])
            df.insert(loc, name, new_col_val)
        elif position["deet"] == "beg":
            df.insert(0, name, new_col_val)
        elif position["deet"] == "end":
            df[name] = new_col_val
        
        return {"val" : True, "detail": df}
        
```
